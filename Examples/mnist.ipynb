```python
from keras.datasets import mnist
import numpy as np
from keras.utils import to_categorical
import sys
sys.path.append('../')
```

```python
from Layers.dense import Dense
from Functions.activation_functions import *
from Functions.loss_functions import MSE
from network import Network
```

```python
def preprocess_data(x, y, limit):
    # reshape input data
    x = x.reshape(x.shape[0], 28 * 28, 1)
    # converting array to float and normalising to avoid large gradient values
    x = x.astype(float) / 255
    # one-hot encode output
    y = to_categorical(y)
    y = y.reshape(y.shape[0], 10, 1)
    return x[:limit], y[:limit]

# load MNIST from server
(train_x, train_y), (test_x, test_y) = mnist.load_data()

print(train_x.shape)
print(train_y.shape)

train_x, train_y = preprocess_data(train_x, train_y, 5000)
test_x, test_y = preprocess_data(test_x, test_y, 50)

print(train_x.shape)
print(train_y.shape)

network = Network([
    Dense(28 * 28, 35),
    Logistic(),
    Dense(35, 10),
    Tanh()
])

# train
network.train(MSE, train_x, train_y, epochs=100, learning_rate=0.1, verbose=True)

# test
stat = 0
for x, y in zip(test_x, test_y):
    output = network.predict(x)
    print('predicted:', np.argmax(output), '\ttrue:', np.argmax(y))
    if np.argmax(output) == np.argmax(y): stat += 1

print(stat/test_x.shape[0] * 100, '%', " successful classifications")
```
