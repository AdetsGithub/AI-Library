```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import sys
import operator
sys.path.append('../')
```

```python
from Functions.activation_functions import *
from Functions.loss_functions import *
from Layers.dense import Dense
from network import Network
```

```python
X = np.reshape([[0, 0], [0, 1], [1, 0], [1, 1]], (4, 2, 1))
Y = np.reshape([[0], [1], [1], [0]], (4, 1, 1))
```

```python
 network = Network([
    Dense(2, 3),
    RelU(),
    Dense(3, 1),
    Tanh()
])
```

```python
# Plotting the xor function
points = []
for x in np.linspace(0, 20, 20):
    for y in np.linspace(0, 20, 20):
        if (x + y) % 2:
            z = 1
        else:
            z = 0
        points.append([x, y, z])

points = np.array(points)

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
ax.scatter(points[:, 0], points[:, 1], points[:, 2], c=points[:, 2], cmap="winter")
plt.show()

# Plotting the predicted output per epoch
for i in range(0, 21):

    network.train(BinaryCrossEntropy, X, Y, epochs=i, learning_rate=0.1, verbose=False)
    
    if i % 1 == 0:
        print("Epoch: ", i)
        points = []
        for x in np.linspace(0, 20, 20):
            for y in np.linspace(0, 20, 20):
                z = network.predict([[x], [y]])
                points.append([x, y, z[0, 0]])

        points = np.array(points)

        fig = plt.figure()
        ax = fig.add_subplot(111, projection="3d")
        ax.scatter(points[:, 0], points[:, 1], points[:, 2], c=points[:, 2], cmap="winter")
        plt.show()
```
