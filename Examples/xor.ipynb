```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import sys
sys.path.append('../')
```

```python
from Functions.activation_functions import Tanh
from Functions.loss_functions import *
from Layers.dense import Dense
from network import Network
```

```python
X = np.reshape([[0, 0], [0, 1], [1, 0], [1, 1]], (4, 2, 1))
Y = np.reshape([[0], [1], [1], [0]], (4, 1, 1))
```

```python
 network = Network([
    Dense(2, 3),
    Tanh(),
    Dense(3, 1),
    Tanh()
])

# training the network
network.train(BinaryCrossEntropy, X, Y, epochs=10000, learning_rate=0.1, verbose=False)
```

```python
# Plotting the xor function
points = []
for x in np.linspace(0, 1, 20):
    for y in np.linspace(0, 1, 20):
        z = 1 if (x+z) == 1 else 0
        points.append([x, y, z])

points = np.array(points)

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
ax.scatter(points[:, 0], points[:, 1], points[:, 2], c=points[:, 2], cmap="winter")
plt.show()

# plotting the decision boundary
points = []
for x in np.linspace(0, 1, 20):
    for y in np.linspace(0, 1, 20):
        z = network.predict([[x], [y]])
        points.append([x, y, z[0, 0]])

points = np.array(points)

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
ax.scatter(points[:, 0], points[:, 1], points[:, 2], c=points[:, 2], cmap="winter")
plt.show()
```
